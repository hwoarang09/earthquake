1년 단위로 그려서 특정 년도에 어떤 일이 발생했는지는 확인하기 쉬운데,<br>
5년 단위로 묶어서도 확인해봅시다.<br>


```python
# 이번에 사용한 pd.cut은 약간 까다로운 코드이니까 천천히 분석해도 됨.
df['Year_5년 단위'] = pd.cut(df.index.year,
                          bins=range(df.index.year.min(), df.index.year.max() + 6, 5),
                          right=False,
                          labels=[f"{i}-{i+4}" for i in range(df.index.year.min(), df.index.year.max(), 5)])
df
```

```python

fig = px.histogram(df, x="Month", color="Year_5년 단위",
                   category_orders={"Month": ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']},
                   barmode='stack',
                  )

fig.update_layout(width=800, height=600)


fig.show()
```

5년 단위로 묶으니까 또 다른 인사이트가 도출되지 않나요?<br><br>

명확하게 하기 위해 년도 별로 다시 그려봅시다.

```python
# x축에 들어갈 연도를 나타내는 list생성
Year = [x for x in range(1978,2025,1)]
```

```python
# 각 연도를 groupby로 그룹화 합니다.
# 다른 것들이랑 다르게 groupby는 출력해도 뭐 이상한 거만 출력됩니다.
yearly_groups = df.groupby(df.index.year)
yearly_groups
```

```python
yearly_groups['규모'].count()
```

```python
# 각 그룹에서 '규모' 열의 개수가 몇개인지 세어서 지진 발생 수 계산
# 다른 열 세도 됨.
Earthquake_count = yearly_groups['규모'].count().values
Earthquake_count
```

```python
def fetch_quakes_data(eqdf):

    center_lat, center_lon = [37.566, 126.9780]
    eqdf['Time'] = pd.to_datetime(eqdf['Time'])
    eqdf = eqdf.sort_values(by='Time')

    return eqdf, center_lat, center_lon
```

```python
def fetch_quakes_data(eqdf):

    center_lat, center_lon = [37.566, 126.9780]
    eqdf['Time'] = pd.to_datetime(eqdf['Time'])
    eqdf = eqdf.sort_values(by='Time')

    return eqdf, center_lat, center_lon

def visualize_quakes_data(df):
    eqdf, clat, clon = fetch_quakes_data(df)


    zoom = 5
    year_i = 1978

    fig = px.scatter_mapbox(
        data_frame=eqdf,
        lat='latitude',
        lon='longitude',
        center=dict(lat=clat, lon=clon),
        size='size',
        color='규모',
        range_color=(0, 6),
        zoom=zoom,
        mapbox_style='carto-darkmatter',
        color_continuous_scale=px.colors.cyclical.IceFire,
        #color_continuous_scale=px.colors.sequential.amp,
        animation_frame='Year',
        title='{} 지진 발생지역 - {}년도'.format("년도별", year_i)
    )

    fig.update_layout(
        margin=dict(l=20, r=0, t=65, b=10)
    )
    fig.update_layout(width=700, height=700)
    fig.update_layout(
        paper_bgcolor='black',  # 플롯 영역의 배경색
        font_color='lightgray'
    )

    for frame in fig.frames:
      year_i+=1
      frame['layout']['title'] = '{} 지진 발생지역 - {}년도'.format("년도별", year_i)

    fig['layout']['sliders'][0]['pad']=dict(l=30, t=10, b=20)
    fig['layout']['updatemenus'][0]['pad']=dict(l=30, t=10, b=20)
    fig.show()
    return None

```

```python
visualize_quakes_data(df)
```
